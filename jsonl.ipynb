### Code to convert pandas dataframe to json line format
This code is used to convert pandas table to json line.
This line format is used as input for DeepAR algorith,
In this format,each json is used for each row of table and no parent curly braces are present in whole file.

```python
import pandas as pd
df = pd.read_csv("Lancet.csv")
df.head()
```

```python
jsonl = df.to_json(orient = 'records', lines = True)
print(jsonl)
```

```python
file1 = open('my-jsonl-file.txt', 'w') 
```

```python
file1.write(jsonl)
file1.close()
```
